```python
def load_data(path,sheet_name):
    '''
    This function loads the file and return the raw data
    : path - Path of the excel file
    : sheet - sheet name of the file where the data is located
    : df - returns a pandas dataframe
    '''
    
    df=pd.read_excel(path,sheet_name=sheet_name)

    return df
```

```python
def fill_nulls(df):
    
    
    df['rcvval'] = df['rcvval'].fillna(0)
    df['orderval'] = df['orderval'].fillna(0)    
    
    return df
```

```python
def create_day_bucket_col(df):

    df['day_bucket'] = df['date'] - pd.datetime(2020,10,22)
    df['day_bucket'] = df['day_bucket'].astype(str)
    df['day_bucket'] = df['day_bucket'].str.slice(0,2)
    
    return df
```

```python
def creat_mapping(df):
    
    #converting the column from str to int
    df['day_bucket'] = df['day_bucket'].astype(int)
    
    #extracting the minimum values
    min_val = df['day_bucket'].value_counts().sort_index().index[0]
    
    #extractin the maximum value
    max_val = df['day_bucket'].value_counts().sort_index().index[-1]
    
    map_dict = {}
    for i in range(min_val,(max_val + 1)):

        if i<0:
            map_dict[i] = '-1 to -10'
        elif i<3:
            map_dict[i] = '0 to 2'
        else:
            map_dict[i] = '3 to 7'

    df['day_bucket']=df['day_bucket'].map(map_dict)
    
    return df
```

```python
def create_groups(df):
    
    d1 = df[df["day_bucket"] == '-1 to -10'].groupby(['format','division_name'])['orderval','rcvval'].sum()
    d2 = df[df["day_bucket"] == '0 to 2'].groupby(['format','division_name'])['orderval','rcvval'].sum()
    d3 = df[df["day_bucket"] == '3 to 7'].groupby(['format','division_name'])['orderval','rcvval'].sum()
    
    d1['fillrate']=d1['rcvval']/d1['orderval']
    d2['fillrate']=d2['rcvval']/d2['orderval']
    d3['fillrate']=d3['rcvval']/d3['orderval']
    
    dflist=[d1,d2,d3]
    df_final=pd.concat(dflist,axis=1)
    df_final
    
    return df_final
```

```python
path = 'Interview_questions.xlsx'
sheet_name = 'base_data'
```

```python
df = load_data(path,sheet_name)
df = fill_nulls(df)
df = create_day_bucket_col(df)
```

```python
df = creat_mapping(df)
df = create_groups(df)
df
```
