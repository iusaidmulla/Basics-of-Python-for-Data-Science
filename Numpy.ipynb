```python
#lets declare a size
size = 100000000

#create two lists
list1 = range(size)
list2 = range(size)

#create two numpy arrays
arr1 = np.arange(size)
arr2 = np.arange(size)
```

```python
#calculate time for python list
start = time.time()
result1 = [(x+y) for x,y in zip(list1,list2)]
print('Time for python lists in msec : ',(time.time() - start))

#calculting for numpy array
start = time.time()
result2 = (arr1 + arr2)
print('Time for numpy array in msec : ',(time.time() - start))
```

#### 6. Find the common and uncommon elements

```python
arr1 = np.array(range(7))
arr2 = np.array(range(5,10))
```

```python
# Print only the common numbers
# Print only the uncommon numbers
print(arr1)
print(arr2)
```

```python
#common element
np.intersect1d(arr1,arr2)
```

```python
np.union1d(arr1,arr2)
```

```python
# uncommon values
np.setdiff1d(np.union1d(arr1,arr2),np.intersect1d(arr1,arr2))
```
